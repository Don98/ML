Analysing data and get the data what we want in the dataset

Step 1:import all packages we need in the programme

```python
import numpy as np
import pandas as pd
```

Step 2: get useful data in original files

```python
df=pd.read_csv('train.csv')
df.head()
```

```python
df1=df[df['observation']=='PM2.5']
df1
```

```python
PMdata=df1.iloc[:,3:]
PMdata
```

###### Choose continuous 9 columns as feature which used to be trained,and the tenth as label;
###### If we choose continuous 10 columns randomly, there will be  24-10+1=15 kinds in total

```python
train_x=[]
train_y=[]
for i in range(15):
    x=PMdata.iloc[:,i:i+9]
    x.columns=np.array(range(9))  #notice if we don't set columns name, it will have different columns name in each iteration
    y=PMdata.iloc[:,i+9]
    y.columns=np.array(range(1))
    train_x.append(x)
    train_y.append(y)
```

```python
train_x=pd.concat(train_x)
train_y=pd.concat(train_y)  #train_x and train_y are the type of Dataframe
train_x
```

Change data type to ndarray

```python
# the original elemnt in dataset is str, now we need to change it to float 
train_x=np.array(train_x,float)
train_y=np.array(train_y,float)
train_x
```

###### Train data by using gradient descent
y=wx+b

```python
#add bias to train dataset
train_x=np.concatenate((np.ones((train_x.shape[0],1)),train_x),axis=1)
train_x
```

Gradient descent starts

![steps](img/code1.PNG)

Initialize weight ,learning rate,iteration times

```python
w=np.zeros(len(train_x[0]))
lr=10
Iteration=10000
sum_gra=np.zeros(len(train_x[0]))  #define sum=0
for i in range(Iteration):
    y_new=np.dot(train_x,w)
    loss=y_new-train_y
    gra= 2*np.dot(train_x.transpose(),loss) #notice we must transpose train_x 
    sum_gra += gra**2
    ada=np.sqrt(sum_gra)
    w=w-lr*gra/ada
w    
```

```python
df_test=pd.read_csv('test.csv')
df_test.head()
```

```python
df2=df_test[df_test['AMB_TEMP']=='PM2.5']
df2
```

```python
PMdata_test=df2.iloc[:,2:]
PMdata_test
```

```python
test_x=np.array(PMdata_test,float)
test_x
```

Add bias to test data

```python
test_x=np.concatenate((np.ones((test_x.shape[0],1)),test_x),axis=1)
test_x
```

```python
y_pre=np.dot(test_x,w)
y_pre
```

```python
y_submit=pd.read_csv('sampleSubmission.csv')
y_submit.value=y_pre   #get value by .value   automatically identify numbers
y_submit.to_csv('sampleSubmission.csv',index=False)
```

Calculate error

```python
real=pd.read_csv('ans.csv')
error=abs(y_submit.value-real.value).sum()/len(real.value)
error
```
